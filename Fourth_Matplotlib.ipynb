```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100000)
print(x)
y = np.sin(x)
print(y)
plt.plot(x,y)
```

```python
x = np.linspace(0,10,100)
print(x)
y = np.sin(x)
print(y)
plt.scatter(x,y,color='red')
```

```python
plt.plot(x,y,color='b')
plt.plot(x,np.cos(x),color='green')
```
